```python
import pandas as pd
from scipy.stats import ks_2samp

# Load data from CSV
df = pd.read_csv(r'C:\E\Posdoc work at Upenn\Lab_work\Projected_I_am_intersted_in\Transcriptome of human DRG neurons\Manuscript and figs\Submmision\Nature neuroscience\Final\Revision\Data analysis\Dissected soma size distribution\hDRG neurons.csv')

# Extract the two datasets
data1 = df['All_cells_staining'].dropna()  # Dropping any NaN values
data2 = df['All_dissected_cells'].dropna()  # Dropping any NaN values

# Perform two-sample K-S test
D, p_value = ks_2samp(data1, data2)

# Print results
print(f"K-S statistic: {D}")
print(f"p-value: {p_value}")

if p_value < 0.05:
    print("The two samples likely come from different distributions.")
else:
    print("There's no significant difference between the two distributions.")
```
